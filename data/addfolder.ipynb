```python
import os
```

```python
image_folder = 'images/'
label_folder = 'labels/'
for folder in os.listdir(image_folder):
    if 'NW' in folder:
        if not os.path.exists(f'{label_folder}{folder}'):
            os.mkdir(f'{label_folder}{folder}')
            os.mkdir(f'{label_folder}{folder}/labels-empty-YOLO')
        for lab in os.listdir(f'{image_folder}{folder}'):
            name = lab.split('.')[0]
            print(f'{label_folder}{folder}/labels-empty-YOLO/{name}.txt')
            with open(f'{label_folder}{folder}/labels-empty-YOLO/{name}.txt', 'w'): pass
```

```python
# this fills empty anno files
image_folder = 'images/'
label_folder = 'labels/'
for folder in os.listdir(image_folder):
    if 'W' in folder:
        for lab in os.listdir(f'{image_folder}{folder}'):
            name = lab.split('.')[0]
            if not os.path.exists(f'{label_folder}{folder}'): continue
            file = [file for file in os.listdir(f'{label_folder}{folder}') if file.endswith('YOLO')][0]
            print(f'{label_folder}{folder}/{file}/{name}.txt')
            if not os.path.exists(f'{label_folder}{folder}/{file}/{name}.txt'):
                with open(f'{label_folder}{folder}/{file}/{name}.txt', 'w'): pass
```
